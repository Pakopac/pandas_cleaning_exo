```python
import pandas as pd
```

```python
url = "https://bit.ly/missing-values"
df = pd.read_csv(url)
```

```python
df['color_missing'] = df['color'].isna()
```

```python
df.head()
```

```python
df.insert(0, "color_missing2", df['color'].isna())
```

```python
df.head()
```

## Insérer une colonne pour renseigner la présence de valeur manquante [NUMERIQUE / ORDINAL]

```python
pd.get_dummies(df, dummy_na=True)
```

```python
# ici n'incluer que le nom des colonnes NUMERIQUE/ORDINAL
for nom_colonne in ['color', 'rating']:
    index_colonne = df.columns.get_loc(nom_colonne)
    df.insert(index_colonne+1,nom_colonne+'_nan',df[nom_colonne].isna().astype('int8'))
```

```python
df.head()
```

La touche finale pour les colonnes ordinales,  colonne.cat.codes

La touche finale pour les colonnes numériques, as_type("int") ou as_type("float")

```python
df['color'].isna().sum()
```

```python
df['rating'].fillna(df['rating'].mean())
```

```python
df['color'].value_counts()
```

```python
df['color'].fillna('Manquante')
```

```python
df['color'] = df['color'].astype('category')
df.dtypes
```

```python
pd.get_dummies(df, dummy_na=True, columns=['color']).head()
```

```python
df.to_csv('nom_fichier.csv', index=None)
```

Any enlève la ligne si une des valeurs des colonnes sélectionnées est NaN

```python
df2.dropna(how="all",subset=['Colonne Importante'])
```

```python
df2.dropna(how="any",subset=['Colonne Importante', 'Colonne Peu Utile'])
```
